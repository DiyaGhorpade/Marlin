```python
import requests
from bs4 import BeautifulSoup
import re
import urllib.request
import pdfplumber
import pandas as pd
```

```python
url = r"https://www.cmfri.org.in/data-publications"
```

```python
response = requests.get(url)
```

```python
response
```

```python
soup= BeautifulSoup(response.text, "html.parser")     
```

```python
links = [
    a['href']
    for a in soup.find_all('a', href=True)
    if a['href'].lower().endswith('.pdf') and ('fish' and 'landing' in a['href'].lower())
]
```

```python
links
```

```python
filenames = []
for link in links:
    filename = link.split("/")[-1]  # extract filename
    print(f"Downloading {filename} ...")
    filenames.append(filename)
    # response = requests.get(link)
    # with open(filename, "wb") as f:
        # f.write(response.content)
print("All downloads complete.")
```

```python
filenames
```

```python
tables = []
text_data = []
```

```python
for filename in filenames:
    with pdfplumber.open(filename) as pdf:
        for page in pdf.pages:
            # Extract any tables
            tables_on_page = page.extract_tables()
            for table in tables_on_page:
                df = pd.DataFrame(table)
                tables.append(df)
            
            # Extract text for NLP later
            text_data.append(page.extract_text())

print(len(tables), "tables extracted")
```
